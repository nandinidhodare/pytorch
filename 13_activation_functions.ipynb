```python
import torch
import torch.nn as nn
import torch.nn.functional as F
```

```python
x = torch.tensor([-1.0,1.0,2.0,3.0])
```

```python
# softmax
output = torch.softmax(x,dim=0)
```

```python
print(output)
```

```python
print(output)
```

```python
sm = nn.Softmax(dim=0)
output = sm(x)
print(output)
```

```python
# sigmoid
output = torch.sigmoid(x)
print(output)
s = nn.Sigmoid()
output = s(x)
print(output)
```

```python
# tanh
output = torch.tanh(x)
print(output)
t = nn.Tanh()
output = t(x)
print(output)
```

```python
# relu
output = torch.relu(x)
print(output)
r = nn.ReLU()
output = r(x)
print(output)
```

```python
#leaky relu
output = F.leaky_relu(x)
print(output)
lr = nn.LeakyReLU()
output = lr(x)
print(output)
```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style('whitegrid')
```

```python
# Sigmoid
sigmoid = lambda x: 1/(1+np.exp(-x))
x = np.linspace(-10,10,10)
y = np.linspace(-10,10,100)
fig = plt.figure()
plt.plot(y,sigmoid(y),'b',label='linspace(-10,10,100)')
plt.grid(ls='--')
plt.xlabel("X axis")
plt.ylabel("Y axis")
plt.title('Sigmoid func')
plt.xticks([-4,-3,-2,-1,0,1,2,3,4])
plt.yticks([-2,-1,0,1,2])
plt.ylim(-2,2)
plt.xlim(-4,4)
plt.show()
```

```python
# tanh
tanh = lambda x: 2*sigmoid(2*x)-1
x = np.linspace(-10,10,10)
y = np.linspace(-10,10,100)
fig = plt.figure()
plt.plot(y,tanh(y),'b',label='linspace(-10,10,100)')
plt.grid(ls='--')
plt.xlabel("X axis")
plt.ylabel("Y axis")
plt.title('tanh func')
plt.xticks([-4,-3,-2,-1,0,1,2,3,4])
plt.yticks([-2,-1,0,1,2])
plt.ylim(-2,2)
plt.xlim(-4,4)
plt.show()
```

```python
# Relu
relu = lambda x: np.where(x>=0,x,0)
x = np.linspace(-10,10,10)
y = np.linspace(-10,10,100)
fig = plt.figure()
plt.plot(y,relu(y),'b',label='linspace(-10,10,100)')
plt.grid(ls='--')
plt.xlabel("X axis")
plt.ylabel("Y axis")
plt.title('Sigmoid func')
plt.xticks([-4,-3,-2,-1,0,1,2,3,4])
plt.yticks([-2,-1,0,1,2])
plt.ylim(-2,2)
plt.xlim(-4,4)
plt.show()
```

```python
# leaky Relu
leakyrelu = lambda x: np.where(x>=0,x,0.1*x)
x = np.linspace(-10,10,10)
y = np.linspace(-10,10,100)
fig = plt.figure()
plt.plot(y,relu(y),'b',label='linspace(-10,10,100)')
plt.grid(ls='--')
plt.xlabel("X axis")
plt.ylabel("Y axis")
plt.title('Sigmoid func')
plt.xticks([-4,-3,-2,-1,0,1,2,3,4])
plt.yticks([-2,-1,0,1,2])
plt.ylim(-2,2)
plt.xlim(-4,4)
plt.show()
```
